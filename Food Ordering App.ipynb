```python
food_items = []

def generate_food_id():
    if len(food_items) == 0:
        return 1
    else:
        max_id = max(food_item['FoodID'] for food_item in food_items)
        return max_id + 1

def add_food_item(name, quantity, price, discount, stock):
    food_id = generate_food_id()
    food_item = {
        'FoodID': food_id,
        'Name': name,
        'Quantity': quantity,
        'Price': price,
        'Discount': discount,
        'Stock': stock
    }
    food_items.append(food_item)
    print(f"Food item with FoodID {food_id} has been added.")

def edit_food_item(food_id, name=None, quantity=None, price=None, discount=None, stock=None):
    for food_item in food_items:
        if food_item['FoodID'] == food_id:
            if name:
                food_item['Name'] = name
            if quantity:
                food_item['Quantity'] = quantity
            if price:
                food_item['Price'] = price
            if discount:
                food_item['Discount'] = discount
            if stock:
                food_item['Stock'] = stock
            print(f"Food item with FoodID {food_id} has been updated.")
            return
    print(f"Food item with FoodID {food_id} not found.")

def view_food_items():
    if len(food_items) == 0:
        print("No food items found.")
    else:
        for food_item in food_items:
            print(f"FoodID: {food_item['FoodID']}")
            print(f"Name: {food_item['Name']}")
            print(f"Quantity: {food_item['Quantity']}")
            print(f"Price: {food_item['Price']}")
            print(f"Discount: {food_item['Discount']}")
            print(f"Stock: {food_item['Stock']}")
            print("---------------------------")

def remove_food_item(food_id):
    for food_item in food_items:
        if food_item['FoodID'] == food_id:
            food_items.remove(food_item)
            print(f"Food item with FoodID {food_id} has been removed.")
            return
    print(f"Food item with FoodID {food_id} not found.")

add_food_item("Pizza", "1 piece", 10.99, 0.1, 20)
add_food_item("Burger", "1 piece", 5.99, 0.0, 15)

view_food_items()

edit_food_item(1, price=12.99, stock=10)

view_food_items()

remove_food_item(2)

view_food_items()


user_profiles = []

def register_user(full_name, phone_number, email, address, password):
    user_profile = {
        'Full Name': full_name,
        'Phone Number': phone_number,
        'Email': email,
        'Address': address,
        'Password': password
    }
    user_profiles.append(user_profile)
    print("User registered successfully.")

def login_user(email, password):
    for user_profile in user_profiles:
        if user_profile['Email'] == email and user_profile['Password'] == password:
            print("Login successful.")
            return
    print("Invalid email or password.")

def display_user_options():
    print("Options:")
    print("1. Place New Order")
    print("2. Order History")
    print("3. Update Profile")

def place_new_order():
    food_menu = [
        {'Name': 'Tandoori Chicken', 'Quantity': '4 pieces', 'Price': 240},
        {'Name': 'Vegan Burger', 'Quantity': '1 piece', 'Price': 320},
        {'Name': 'Truffle Cake', 'Quantity': '500gm', 'Price': 900}
    ]
    print("Food Menu:")
    for i, food_item in enumerate(food_menu):
        print(f"{i + 1}. {food_item['Name']} ({food_item['Quantity']}) [INR {food_item['Price']}]")
    
    selected_items = input("Enter the numbers of the items you want to order, separated by commas: ")
    selected_items = [int(item_num.strip()) for item_num in selected_items.split(',')]

    order_items = []
    total_price = 0
    for item_num in selected_items:
        if 1 <= item_num <= len(food_menu):
            item = food_menu[item_num - 1]
            order_items.append(item)
            total_price += item['Price']

    print("Selected Items:")
    for item in order_items:
        print(f"{item['Name']} ({item['Quantity']}) [INR {item['Price']}]")

    place_order = input("Do you want to place the order? (yes/no): ")
    if place_order.lower() == 'yes':
        # Logic to place the order
        print("Order placed successfully.")
    else:
        print("Order canceled.")

def view_order_history():
    # Logic to retrieve and display order history
    print("Order History:")
    # Display order history information here

def update_profile(email, password):
    for user_profile in user_profiles:
        if user_profile['Email'] == email and user_profile['Password'] == password:
            # Update user profile fields
            full_name = input("Enter Full Name: ")
            phone_number = input("Enter Phone Number: ")
            address = input("Enter Address: ")

            user_profile['Full Name'] = full_name
            user_profile['Phone Number'] = phone_number
            user_profile['Address'] = address

            print("Profile updated successfully.")
            return
    print("Invalid email or password.")

register_user("John Doe", "1234567890", "johndoe@example.com", "123 Main St", "password123")
login_user("johndoe@example.com", "password123")

display_user_options()

option = int(input("Enter the option number: "))
if option == 1:
    place_new_order()
elif option == 2:
    view_order_history()
elif option == 3:
    email = input("Enter your email: ")
    password = input("Enter your password: ")
    update_profile(email, password)
else:
    print("Invalid option number.")
```
