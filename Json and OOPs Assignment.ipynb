```python
import json

employees = [
    {
        "Name": "John Doe",
        "DOB": "1990-05-15",
        "Height": 175,
        "City": "New York",
        "State": "New York"
    },
    {
        "Name": "Jane Smith",
        "DOB": "1985-09-20",
        "Height": 165,
        "City": "Los Angeles",
        "State": "California"
    },
    {
        "Name": "Michael Johnson",
        "DOB": "1992-02-10",
        "Height": 180,
        "City": "Chicago",
        "State": "Illinois"
    },
    {
        "Name": "Emily Davis",
        "DOB": "1988-07-08",
        "Height": 170,
        "City": "Houston",
        "State": "Texas"
    },
    {
        "Name": "Robert Wilson",
        "DOB": "1994-11-25",
        "Height": 185,
        "City": "Miami",
        "State": "Florida"
    }
]

with open('employee.json', 'w') as f:
    json.dump(employees, f)
```

```python
class Employee:
    def __init__(self, name, dob, height, city, state):
        self.name = name
        self.dob = dob
        self.height = height
        self.city = city
        self.state = state

with open('employee.json') as f:
    data = json.load(f)

employees = []

for emp_data in data:
    employee = Employee(emp_data['Name'], emp_data['DOB'], emp_data['Height'], emp_data['City'], emp_data['State'])
    employees.append(employee)

for employee in employees:
    print(f"Name: {employee.name}")
    print(f"DOB: {employee.dob}")
    print(f"Height: {employee.height}")
    print(f"City: {employee.city}")
    print(f"State: {employee.state}")
    print()
```

```python
class Dog:
    def __init__(self, name, age, coat_color):
        self.name = name
        self.age = age
        self.coat_color = coat_color

    def description(self):
        print(f"Name: {self.name}")
        print(f"Age: {self.age}")

    def get_info(self):
        print(f"Coat Color: {self.coat_color}")

class JackRussellTerrier(Dog):
    def bark(self):
        print("Jack Russell Terrier is barking!")

    def play_fetch(self):
        print("Jack Russell Terrier is playing fetch!")

class Bulldog(Dog):
    def snore(self):
        print("Bulldog is snoring!")

    def drool(self):
        print("Bulldog is drooling!")

dog1 = Dog("Buddy", 5, "Brown")
dog1.description()
dog1.get_info()
print()

terrier = JackRussellTerrier("Max", 3, "White and Brown")
terrier.description()
terrier.get_info()
terrier.bark()
terrier.play_fetch()
print()

bulldog = Bulldog("Rocky", 4, "White")
bulldog.description()
bulldog.get_info()
bulldog.snore()
bulldog.drool()
```
